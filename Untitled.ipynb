```python
# importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Get the dimensions (number of rows and columns)
num_rows, num_cols = df.shape

# Print the dimensions
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
#Display the first 3 rows
df.head(3)
```

```python
# Select columns 1 to 11 and convert to a NumPy matrix
df_eleven = df.iloc[:, 0:11].values
print(df_eleven)

# Convert "Yes" to 1 and "No" to 0
df_eleven = (df_eleven == "Yes").astype(int)
print(df_eleven)

# Calculate and round column means
col_means = np.round(np.mean(df_eleven, axis=0), 2)

# # Print the rounded column means
print(col_means)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Standardize the data
scaler = StandardScaler()
df_eleven_std = scaler.fit_transform(df_eleven)

pca = PCA(n_components=11)
pc = pca.fit_transform(df_eleven_std)
names=["pc1", "pc2", "pc3", "pc4", "pc5", "pc6", "pc7", "pc8", "pc9", "pc10", "pc11"]
pf = pd.DataFrame(data=pc, columns=names)
print(pf)

#Summary of PCA results
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:")
print(explained_variance_ratio)
```

```python
std_dev = []
for i in names:
    std_dev.append(np.std(pf[i]))
    
print("Standard Deviation")
np.array(std_dev)
```

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(pf)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title("Elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
#K-means clustering 
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(df_eleven)
df['cluster_num'] = kmeans.labels_ #adding to df
print (kmeans.labels_) #Label assigned for each data point
print (kmeans.inertia_) #gives within-cluster sum of squares. 
print(kmeans.n_iter_) #number of iterations that k-means algorithm runs to get a minimum within-cluster sum of squares
print(kmeans.cluster_centers_) #Location of the centroids on each cluster. 
```
